```python
# import pandas package
import pandas as pd
```

```python
# import numpy
import numpy as np
```

```python
#read thedata set and store it in football
football=pd.read_csv('Laliga.csv')
```

```python
#Check if all the data is correctly loaded
football.columns
```

```python
#replacing all the dashes with 0, for question 1
football=football.replace(to_replace ="-", value ="0")
```

```python
# verifying for all the dashesbeing correctly replaced
football
```

```python
#Printing all the teams that had a debyted between 1930 and 1980 get displaced along with the Team name, for question 2
for i in range(0,football.Debut.count()):
	if (1930<=int(str(football.Debut[i])[:4])<=1980):
		print (football.Team[i], football.Debut[i])
```

```python
#Converting datatype of Points to int
football.Points=football.Points.astype(str).astype(int)
```

```python
#Question 3 to display top 5 team having highest Points
football[['Points','Team']].sort_values(by = 'Points',ascending = False).head(5)
```

```python
#Creating function to calculate and print goal difference 
def Goal_diff_count():   
     football['Goal_diff_count'] = football['GoalsFor'].astype(int)-football['GoalsAgainst'].astype(int)
     return football[['Team','Goal_diff_count']].sort_values(by = 'Goal_diff_count',ascending=False)
Goal_diff_count()
```

```python
# Team with Maximun goal diffence, question 4
print("The team having maximum goal difference is :")
Goal_diff_count().head(1)
```

```python
# Team with minimum goal diffence, question 4
print("The team having minimum goal difference is :")
Goal_diff_count().tail(1)
```

```python
#Calculate win percentage
football['Winning Percent'] = (football['GamesWon'].astype(int)/football['GamesPlayed'].astype(int)) *100
```

```python
#If any numerical error replace it with 0, question 5
football['Winning Percent'].fillna(0,inplace = True)
```
